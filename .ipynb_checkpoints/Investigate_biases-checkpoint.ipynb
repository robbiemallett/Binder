```python
# Import some standard modules
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from math import sqrt
import pandas as pd
import matplotlib as mpl
import pickle

# Import custom modules
import propagation
import Tools
import Mask

# Set some font sizes

figwidth = 12
figheight = 6
annotation_fontsize = 12

# Import data on where MYI usually is (for figures)

pickle_in = open("myiprobEASE","rb")
myidict = pickle.load(pickle_in)

# Import a mask that allows us to focus analysis on the Central Arctic basin
# This mask also helps us work out the rate of snow densification in different regions

mymask = Mask.EASE()

# Create some dicts to iterate over

month_annotations = {"10":'Oct',"11":'Nov',"12":'Dec',"01":'Jan',"02":'Feb',"03":'Mar',"04":'Apr'}

monthlist = ["10", "11", "12", "01", "02", "03", "04"]

yearlist = ["2010", "2011", "2012", "2013", "2014",
            "2015", "2016", "2017", "2018"]
```

## Create Figure 1a & 1b

```python
#############################

# CREATE FIG 1a

#############################

#Create a range of values over which to calculate correct and incorrect propagation corrections

rho = np.linspace(0.2,0.5,101)

#Initiate some empty lists in which to put the data

rightprop, wrongprop, error = [], [], []

#Now populate these lists with the propagation corrections and the difference

for i in rho:

    rightprop.append(propagation.factor(i,"Ulaby"))
    wrongprop.append(propagation.wrong_factor(i,"Ulaby"))
    
    # Calculate the difference between the two corrections
    
    error.append((propagation.factor(i,"Ulaby")-propagation.wrong_factor(i,"Ulaby")))

fig = plt.figure(figsize=(figwidth,figheight-1))

#Plot the three lines  

ax = plt.subplot(1,2,1)

ax.plot(rho,rightprop,color='#984ea3',label=r"c/c$_s$ - 1")

ax.plot(rho,wrongprop,color='#ff7f00',label="1 - c$_s$/c")

ax.plot(rho,error,'k-.',label="Difference, $B_f / Z$",)

ax.set_ylabel('$f(c_s)$',fontsize='medium')
ax.set_xlabel('Density of Dry Snow (g/cm$^3$)', fontsize=annotation_fontsize+5)
ax.set_ylim([0,0.52])
ax.legend(fontsize=annotation_fontsize+5)
ax.annotate('(a)',xy=(-0.2,1.03),xycoords=ax,fontsize=annotation_fontsize+10)

# The following line creates the '(c)' annotation for Fig 1c

ax.annotate('(c)',xy=(-0.2,-0.25),xycoords=ax,fontsize=annotation_fontsize+10)

#############################

# CREATE FIG 1b

#############################

# Initialise an empty dictionary to hold our four datasets

Depth_dict={}

for SWE in [5,10,15]:
    Depth_dict[SWE] = []
    for i in range(len(rho)):

        Depth_dict[SWE].append(error[i]*(SWE/rho[i]))

# Plot the four datasets

ax = plt.subplot(1,2,2)

linecolors = ['#377eb8','#4daf4a','#e41a1c']

for key, color in zip(Depth_dict, linecolors):
    ax.plot(rho, np.array(Depth_dict[key])*9.55,label="SWE: "+str(int(key))+"cm",color=color)
    
for key, color in zip(Depth_dict, linecolors):
    ax.plot(rho, np.array(Depth_dict[key])*7.21,linestyle='--',color=color)
    
# Label axes
ax.set_ylabel('SIT Bias, $B_{SIT}$ (cm)',fontsize='medium',)
ax.yaxis.set_label_position("right")

ax.set_xlabel('Density of Dry Snow (g/cm$^3$)', fontsize=annotation_fontsize+5)

ax.yaxis.tick_right()
ax.set_ylim(0,40)

# Create legend and annotate subplot

ax.legend(fontsize=annotation_fontsize+5)

ax.annotate('(b)',xy=(-0.1,1.03),xycoords=ax,fontsize=annotation_fontsize+10)

plt.savefig("Fig1top.png",dpi=600, bbox_inches="tight")

plt.show()
```

```python
def propagation_correction(month,year,
          correct=True,
          density_evolution = True,
          ref_value = 0.22,
          density_type='linear',
          grid_resolution=(348,348)):
    
    snow_depths =  pickle.load( open( "snowdepths_EASE.p", "rb" ) )[int(month)]
    snow_densities =  pickle.load( open( "snowdensities_EASE.p", "rb" ) )[int(month)]
    
    regionmask = Mask.EASE()
    
    ice_type_directory = 'Monthly_means/'
    
    ice_types = pickle.load( open(ice_type_directory + year + month + "monmean_EASE.p", "rb" ) )

    rho_w=1023.9  #Values from Tilling 2018
    c = 3e8
    
    countdict = {"10":0,"11":1,"12":2,"01":3,"02":4,"03":5,"04":6}

    propagation_correction = np.zeros(grid_resolution)
    coefficient_field = np.zeros(grid_resolution)
    
    for i in range(grid_resolution[0]):                     
        for j in range(grid_resolution[1]):

            if ice_types[i,j] not in [2,3]: #do we have sea ice?
                propagation_correction[i,j] = np.nan
                coefficient_field[i,j] = np.nan
            elif regionmask[i,j] < 7 or regionmask[i,j] == 14:
                propagation_correction[i,j] = np.nan
                coefficient_field[i,j] = np.nan
            else:
                if ice_types[i,j] == 2:
                    depth = snow_depths[i,j]/2
                    ice_density = 916.7
                elif ice_types[i,j] == 3:
                    depth = snow_depths[i,j]
                    ice_density = 882
                
                if density_type == 'linear':
                    snow_density = ((countdict[month]*6.46)+275.4)
                elif density_type == 'distributed':
                    snow_density = snow_densities[i,j]

                E_dry = 1 + 1.7*(snow_density/1000) + 0.7*((snow_density/1000)**2)
                
                if E_dry < 0:
                    propagation_correction[i,j] = np.nan
                    coefficient_field[i,j] = np.nan
                else:
                
                    cs = c/sqrt(E_dry)

                    if correct == True:

                        propagation_correction[i,j] = depth*((c/cs)-1)
#                         propagation_correction[i,j] = depth*0.25

                    if correct == False:
#                         propagation_correction[i,j] = depth*0.22
                        propagation_correction[i,j] = depth*(1-(cs/c))

                    if density_evolution == False:

                        propagation_correction[i,j] = depth*ref_value


                    coefficient_field[i,j] = rho_w/(rho_w - ice_density)
                
    return({"propagation correction":propagation_correction,"coefficient field":coefficient_field})
```

```python
def bias_calculator(errorcorrection=True,
                 densitycorrection=True,
                 density_type='linear',
                 grid_resolution=(348,348)
                 ):
    
    return_dict = {}
    
    for month in monthlist:
        count = 0
        freeboard_diff_sum = np.zeros(grid_resolution)
        coeff_sum = np.zeros(grid_resolution)             

        for year in yearlist:

            data1 = propagation_correction(month,year,
                                           correct=True,
                                           density_evolution=True,
                                           density_type=density_type)

            data2 = propagation_correction(month,year,
                                           correct=errorcorrection,
                                           density_evolution=densitycorrection,
                                           density_type=density_type)

            freeboard_difference = data1["propagation correction"]-data2["propagation correction"]

            coefficient_field = data1["coefficient field"]

            freeboard_diff_sum = np.add(freeboard_diff_sum, freeboard_difference)

            coeff_sum = np.add(coeff_sum, coefficient_field)

            count = count+1

        average_freeboard_difference = np.true_divide(freeboard_diff_sum,count)

        average_coeff_sum = np.true_divide(coeff_sum,count)

        average_thickness_difference = average_freeboard_difference * average_coeff_sum

        return_dict[month] = {"freeboard difference":average_freeboard_difference,
                              "thickness difference":average_thickness_difference}
    
    
    freedifflist, thicdifflist = [],[]
    for key in return_dict:
        freedifflist.append(return_dict[key]['freeboard difference'])
        thicdifflist.append(return_dict[key]['thickness difference'])
    
    return_dict['average freeboard difference'] = np.nanmean(freedifflist,axis=0)
    return_dict['average thickness difference'] = np.nanmean(thicdifflist,axis=0)
    
    
    return(return_dict)
```

```python
radar_freeboards = {}
for month in monthlist:
    print(month)
    sum_array = np.zeros((348,348))
    count = 0
    for year in yearlist:
        data = Tools.getEASE(month,year,"Sea_Ice_Freeboard_W99")
        if data is None:
            print("Failure:"+month+year)
        else:
            sum_array = np.add(data, sum_array)
            count +=1
    radar_freeboards[month] = np.true_divide(sum_array, count)
```

```python
def perc_plots(data,filename):
    percentages = {}
    mean_percentages = {}

    fig = plt.figure(figsize=(figwidth,figheight))

    count = 2

    #general plot stuff

    lon=Tools.getEASE("01", "2015", 'Longitude')
    lat=Tools.getEASE("01", "2015", 'Latitude')

    for month in monthlist:

        percentage = (np.divide(data[month]['freeboard difference'],radar_freeboards[month]))*100
        percentages[month] = percentage
        mean_percentages[month_annotations[month]] = np.nanmean(percentage)

        ax = plt.subplot(2,4,count)

        ax.annotate(month_annotations[month], xycoords=ax,
                         xy=(10, 10),xytext=(0.03, 0.835), fontsize=annotation_fontsize+5)

        m = Basemap(projection='npstere',boundinglat=65,lon_0=360,resolution='l')

        ax = m.drawcoastlines(linewidth=0.3)
        
        levels = np.arange(0,12,2)
        
        ax = m.contourf(lon, lat,
                  percentage,
                  levels,
                  latlon=True,
                  cmap='Reds',
                  vmax=10,
                  vmin=0)

        ax = m.contour(lon, lat, myidict[month], 1, colors='k', linestyles='-',linewidths=0.7,latlon=True)

        count = count + 1


    ####################################################################
    ax = plt.subplot(2,4,1)

    ax.set_ylabel('Mean Bias as % of Ice Freeboard')

    ax.bar(mean_percentages.keys(),mean_percentages.values())

    plt.subplots_adjust(wspace=0.1, hspace=0.2)
    
    plt.savefig(filename, dpi=600, bbox_inches='tight')
```

```python
def hist_plots(output1,outputfname):

    fig = plt.figure(figsize=(figwidth,figheight))

    for month, num in zip(month_annotations, range(1,8)):
        
        area = mymask.flatten()
        age_data = myidict[month].flatten()


        freeboard_bias = output1[month]['freeboard difference'].flatten()
        SIT_bias = output1[month]['thickness difference'].flatten()


        dataframe = pd.DataFrame({'age':age_data,
                                  'area':area,
                                  'freeboard_bias':freeboard_bias,
                                  'thickness_bias':SIT_bias})

        dataframe = dataframe[np.logical_not(np.isnan(dataframe['freeboard_bias']))]

        dataframe = dataframe[dataframe['area'].isin([10,11,12,13,15])]

        data_MYI = dataframe[dataframe['age'] > 0.5]

        data_FYI = dataframe[dataframe['age'] < 0.5]

        myimean = np.mean(data_MYI['thickness_bias'])
        fyimean = np.mean(data_FYI['thickness_bias'])

        print("Month: "+month+' MYI Thickness Bias: '+str(myimean)+'FYI Thickness Bias: '+str(fyimean))

        #PLOT THICKNESS BIAS

        ax = fig.add_subplot(2,4,num)

        ax.hist(data_MYI['thickness_bias']*100, color='r',alpha=0.5,label='Multiyear Ice')
        ax.hist(data_FYI['thickness_bias']*100, alpha=0.5,label='First Year Ice')

        ax.axvline(x=myimean*100, color='r', linestyle='dashed', linewidth=2, ymax=0.7)
        ax.axvline(x=fyimean*100, color='#1f77b4', linestyle='dashed', linewidth=2, ymax=0.7)



        ax.annotate(month_annotations[month],xy=(0.03,0.89),xycoords='axes fraction', fontsize=annotation_fontsize+5)

        ax.set_ylim(0,1600)
        ax.set_xlim(0,16)

        ax.set_xticks([0,4,8,12,12,16])
        
        ax.tick_params(axis="x", labelsize=14)
        
        ax.tick_params(axis="y", labelsize=16)
        
        if int(month) in [10,11,12,1]:
            ax.xaxis.set_major_formatter(plt.NullFormatter())


        if int(month) not in [10,2]:
            ax.yaxis.set_major_formatter(plt.NullFormatter())
        else:
            ax.set_yticks([0,300/0.628,600/0.628,900/0.628])
            y_vals = ax.get_yticks()
            ax.set_yticklabels(['{:3.0f}'.format(x * 628 * 0.001) for x in y_vals])
        if month == "01":
            ax.legend(loc='upper right', prop={'size': 10})

    ####################################################################################

    area = mymask.flatten()

    age_data = np.nanmean(np.array(list(myidict.values())), axis = 0).flatten()

    freeboard_bias = output1['average freeboard difference'].flatten()

    SIT_bias = output1['average thickness difference'].flatten()

    dataframe = pd.DataFrame({'age':age_data,
                              'area':area,
                              'freeboard_bias':freeboard_bias,
                              'thickness_bias':SIT_bias})

    dataframe = dataframe[np.logical_not(np.isnan(dataframe['freeboard_bias']))]

    dataframe = dataframe[dataframe['area'].isin([10,11,12,13,15])]

    data_MYI = dataframe[dataframe['age'] > 0.5]

    data_FYI = dataframe[dataframe['age'] < 0.5]

    myimean = np.mean(data_MYI['thickness_bias'])
    fyimean = np.mean(data_FYI['thickness_bias'])

    print("Ave MYI Thickness Bias: "+str(myimean)+'Ave FYI Thickness Bias: '+str(fyimean))

    ax = fig.add_subplot(2,4,8)

    ax.hist(data_MYI['thickness_bias']*100, color='r',alpha=0.5,label='Multiyear Ice')
    ax.hist(data_FYI['thickness_bias']*100, alpha=0.5,label='First Year Ice')

    ax.axvline(x=myimean*100, color='r', linestyle='dashed', linewidth=2, ymax=0.7)
    ax.axvline(x=fyimean*100, color='#1f77b4', linestyle='dashed', linewidth=2, ymax=0.7)

    ax.annotate('Ave.',xy=(0.03,0.9),xycoords='axes fraction', fontsize=annotation_fontsize+5) ##########

    ax.set_ylim(0,1600)
    ax.set_xlim(0,16)

    ax.set_xticks([0,4,8,12,12,16])
    ax.tick_params(axis="x", labelsize=14)
    ax.yaxis.set_major_formatter(plt.NullFormatter())

    ####################################################################################


    fig.text(0.05, 0.5, 'Sea Ice Area ($10^3$ km$^3$)', ha='center', va='center', rotation='vertical', fontsize=annotation_fontsize+4)
    fig.text(0.5, 0.03, 'Sea Ice Thickness Bias (cm)', ha='center', va='center', fontsize=annotation_fontsize+4)

    plt.subplots_adjust(wspace=0.2, hspace=0.1)
    
#     ax.annotate('(c)',xy=(-0.2,1),xycoords=ax,fontsize=annotation_fontsize+10)

    plt.savefig(outputfname, dpi=600, bbox_inches='tight')

    plt.show()
```
