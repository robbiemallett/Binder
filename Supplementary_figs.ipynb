# This notebook plots the supplementary figures

```python
# Import standard modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pickle
import math

from scipy.stats import linregress
from mpl_toolkits.basemap import Basemap

# Import custom modules

import Mask
import Tools
import propagation
```

# Figure S1

```python
c = 3e8 # Speed of light

cs_list_Tiuri, cs_list_Halli, cs_list_Ulaby = [], [], []

rho = np.linspace(0.1,0.6,101) # Initialise a range of densities to investigate

# Calculate radar wave speed in snow at each density

for i in rho:
    cs_list_Tiuri.append(propagation.speed(i,'Tiuri'))
    cs_list_Halli.append(propagation.speed(i,'Hallikainen'))
    cs_list_Ulaby.append(propagation.speed(i,'Ulaby'))

# Plot the figure

plt.figure(figsize=(8,6))    
plt.plot(rho,cs_list_Tiuri,color='r',label="Tiuri et al. 1984")
plt.plot(rho,cs_list_Halli,color='g',label="Hallikainen et al. 1982")
plt.plot(rho,cs_list_Ulaby,color='b',label="Ulaby et al. 1986")
plt.ylabel('Speed of Light in Snow (x 10$^8$)',fontsize=20)
plt.xlabel('Density of Dry Snow (g/cm$^3$)', fontsize=20)
plt.legend(fontsize=20)
# plt.savefig("figS1",dpi=400, bbox_inches="tight")
plt.show()
plt.close()
```

# Figure S3

```python
# Create array of density values

rho_array = np.linspace(200,400,100)
weight_error_array = []
prop_error_array = []

# Examine the error for small density errors around rho = 300

rho_true = 300

# Calculate the propagation correction for rho = 300 using the Ulaby formula

propagation_true = propagation.factor(rho_true,'Ulaby')

# Calculate the snow weight correction for rho = 300

weight_true = 2.4*rho_true

for rho in rho_array:
    
    # Calculate the propagation correction for the densities in question
    
    propagation_biased = propagation.factor(rho,'Ulaby')
    
    # Calculate the snow weight correction for the densities in question
    
    weight_biased = 2.4*rho
    
    # Calculate the differences between corrections for assumed and real densities

    prop_error_array.append(100*(propagation_biased - propagation_true)/propagation_true)

    weight_error_array.append(100*(weight_biased - weight_true)/weight_true)

# Plot the figure

fig = plt.subplots(figsize=(8,6))

plt.plot(rho_array, prop_error_array, label='Propagation Error')
plt.plot(rho_array, weight_error_array, label='Weight Error')
plt.legend(fontsize='x-large')
plt.xlabel('Dry Snow Density (kg/m$^3$)',fontsize='x-large')
plt.ylabel('% Error as Density Deviates from 300kg/m$^3$',fontsize='x-large')
# plt.savefig("figS3.png",dpi=500, bbox_inches="tight")
plt.show()
```

# Figure S4

```python
path ='Warren_climatology.xlsx'
monthlist = ["10", "11", "12", "01", "02", "03", "04"]

SWE_data = pd.read_excel(path,sheet_name="SWE",index_col=0)
depth_data = pd.read_excel(path,sheet_name="depth",index_col=0)

pole_density = []
pole_depth = []
month_axis = []

for month in monthlist:
    dep0 = depth_data.loc[int(month)]["H0"]
    SWE0 = SWE_data.loc[int(month)]["H0"]
    pole_depth.append(dep0)
    pole_density.append(SWE0*1000/dep0)

stats = linregress(np.array(range(7)),pole_density)
print(f'Snow densifies at {stats[0]} kg/m3 per month, beginning at {stats[1]} kg/m3 in October')

monthlist = ["10", "11", "12", "01", "02", "03" ,"04"]

arealist = {'Beaufort':{'code':13},"Chukchi":{'code':12},'East Siberian':{'code':11},
            'Laptev':{'code':10},"Central Arctic":{'code':15}}

density_fields = {}
area_evolution = {}

mymask = Mask.EASE()

for month in monthlist:
    
    density_fields[month] =  pickle.load( open( "snowdensities_EASE.p", "rb" ) )[int(month)]
    
for area in arealist:
    code = arealist[area]['code']
    monthly_mean_density = []
    for month in density_fields:
        field = density_fields[month]
        density_running = []
        for i in range(348):
            for j in range(348):
                if np.isnan(mymask[i,j]) == False:
                    if int(mymask[i,j]) == code:
                        density_running.append(field[i,j])
        monthly_mean_density.append(np.nanmean(density_running))
    arealist[area]['density_series'] = monthly_mean_density
    
arealist['North Pole'] = {'code':99,'density_series':pole_density}

fig = plt.figure(figsize=(8, 5))
for area in arealist:
    plt.plot(monthlist,arealist[area]['density_series'],label=area,marker='.')
    
plt.xlabel('Month',fontsize='large')
plt.ylabel('W99 Snow Density (kg m$^{-3}$)',fontsize='large')
plt.ylim([240,360])
plt.legend(fontsize='large')
plt.tight_layout()
# plt.savefig('figS4',dpi=500)
plt.show()
```
